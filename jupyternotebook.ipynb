```python
# Sort the DataFrame by 'plot_id' and 'date' in descending order
plot_data = plot_data.sort_values(by=['plot_id', 'date'], ascending=[True, False])

# Drop duplicates, keeping the first occurrence (which will be the latest due to sorting)
plot_data = plot_data.drop_duplicates(subset=['plot_id'], keep='first')

# Reset index for a clean DataFrame
plot_data = plot_data.reset_index(drop=True)

# Display the cleaned data
print(plot_data)
```

Q4

```python
plot_tree_join = pd.merge(tree_data, plot_data, on='plot_id')
plot_tree_join
```

```python
# Calculate the total number of trees per plot
trees_per_plot = plot_tree_join.groupby('plot_id')['tree_count'].sum().reset_index(name='total_trees')
trees_per_plot
```

```python
# Calculate the total planned number of trees per plot
plot_data['planned_trees_total'] = plot_data['planting_plan_per_acre'] * plot_data['area_ac']
plot_data
```

```python
# Join the total trees per plot with the plot_data to get the planned_trees_total
trees_per_plot = pd.merge(trees_per_plot, plot_data[['plot_id', 'planned_trees_total']], on='plot_id')
trees_per_plot
```

### Survival Rate: The ratio of total_trees to planned_trees_total tells us the proportion of the planned trees that actually survived. A ratio of 1 would mean that exactly the number of trees planned to be planted have survived. A ratio less than 1 indicates that fewer trees survived than planned, 

### and a ratio greater than 1 would indicate more trees survived than initially planned (which might be due to errors or over-planting).

```python
# Calculate the number of trees survived per plot
trees_per_plot['num_survived_trees'] = trees_per_plot['total_trees'] / trees_per_plot['planned_trees_total']

print(trees_per_plot[['plot_id', 'total_trees', 'planned_trees_total', 'num_survived_trees']])
```

```python
trees_per_plot['survival_rate'] = trees_per_plot['total_trees'] / trees_per_plot['planned_trees_total']*100

# Display the result
print(trees_per_plot[['plot_id', 'total_trees', 'planned_trees_total', 'survival_rate']])
```
